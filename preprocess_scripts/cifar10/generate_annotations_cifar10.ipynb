```python
import os
import pandas as pd
```

# 生成 train_annotations.csv

```python
train_img_name_list = os.listdir("../../dataset/cifar10/processed/train_data")
train_img_name_list
```

```python
train_img_name_df = pd.DataFrame(train_img_name_list, columns=["name"])
train_img_name_df
```

```python
train_img_name_df["name_label"] = train_img_name_df["name"].map(lambda x: x.split("_")[0])
train_img_name_df["name_label"].value_counts()
```

```python
label_dict = {"airplane": 0, "automobile": 1, "bird": 2, "cat": 3, "deer": 4, "dog": 5, "frog": 6, "horse": 7, "ship": 8, "truck": 9}
train_img_name_df["label"] = train_img_name_df["name_label"].map(label_dict)
train_img_name_df
```

```python
train_img_name_df = train_img_name_df[["name", "label"]]
train_img_name_df
```

```python
train_img_name_df.to_csv("../../dataset/cifar10/processed/full_train_annotations.csv", index=False)
```

# 总训练集拆分成训练集和验证集

```python
from sklearn.model_selection import train_test_split

train_annotations = pd.read_csv("../../dataset/cifar10/processed/full_train_annotations.csv")

train_df, val_df = train_test_split(
    train_annotations,
    test_size=0.2,
    random_state=42,
    stratify=train_annotations["label"],
    # shuffle=False,
)

train_df.to_csv("../../dataset/cifar10/processed/train_annotations.csv", index=False)
val_df.to_csv("../../dataset/cifar10/processed/val_annotations.csv", index=False)
```

```python
# 将验证集数据移动到新的文件夹中：val_data
for item in val_df['name']:
    print(item)
```

```python
import shutil

source_dir = "../../dataset/cifar10/processed/train_data"
target_dir = "../../dataset/cifar10/processed/val_data"
os.makedirs(target_dir, exist_ok=True)

# 将验证集数据移动到新的文件夹中：val_data
for item in val_df['name']:
    # print(item)
    source_img_path = os.path.join(source_dir, item)
    target_img_path = os.path.join(target_dir, item)
    try:
        shutil.move(source_img_path, target_img_path)
    except:
        print(f"failed to move {item}")
```
