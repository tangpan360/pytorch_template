```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# 1. 读取 JSON Lines 文件：每行都是一个 JSON 对象
log_path = "../logs/ag_news/training_metrics.jsonl"

records = []
with open(log_path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line:  # 确保不是空行
            record = json.loads(line)
            records.append(record)

# 2. 将列表转为 DataFrame
df = pd.DataFrame(records)

# 3. 查看前几行，验证结构
print(df.head())

# 4. 创建一个图形对象，包含一个主轴 ax1 用于绘制 loss 曲线
fig, ax1 = plt.subplots(figsize=(8, 6))

# 绘制训练/验证的 loss 曲线
ax1.plot(df["epoch"], df["train_loss"], label="Train Loss", marker='o', color='tab:blue')
ax1.plot(df["epoch"], df["val_loss"], label="Val Loss", marker='o', color='tab:orange')
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Loss", color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.legend(loc='upper left')
ax1.grid(True)

# 创建第二个 y 轴 ax2 用于绘制 accuracy 曲线
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

# 绘制训练/验证的 accuracy 曲线
ax2.plot(df["epoch"], df["train_acc"], label="Train Acc", marker='s', color='tab:green')
ax2.plot(df["epoch"], df["val_acc"], label="Val Acc", marker='s', color='tab:red')
ax2.set_ylabel("Accuracy (%)", color='tab:green')  # we already handled the x-label with ax1
ax2.tick_params(axis='y', labelcolor='tab:green')
ax2.legend(loc='upper right')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title("Training & Validation Loss and Accuracy")
plt.show()
```
